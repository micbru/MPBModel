This notebook is specifically to analyze what happens when we are very close to the equilibrium point, since the system loses stability before regaining it. This is largely for the appendix, and is only for the Hill function approximation. This range is quite small, and therefore unlikely to be relevant biologically.

1. Figure of the spectral radius very close to the critical point.
2. Simulations very close to the critical point.

```python
# Setup
import analysisfunctions as af
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'serif'

# Fixed parameters
# Beetle productivity
c = 1 # Nondimensionalized
# Fixed values of n and theta (theta, not phi, since c =1)
n=5;t=0.4;s=0.99;N=50;
# Find the critical point for m and theta
ms = c*(n-1)/n
teq = (n-1)**((n-1)/n)/n*c

# Colors
c1 = '#00485C' # This is the official dark blue
c2 = '#0084A8' # Official light blue
c3 = '#986532' # Official "yellow", which is more dark brown.
cg = '0.4' # Grey
# Other options:
#'#F2E2CA' # Pale straw yellow 
# And generate an array of lightness values, from 5% to 95% lightness including the official one which is 18%.
# 20 colors, and index 16 is the initial one
# Useful link: https://www.w3schools.com/colors/colors_picker.asp
c_arr = [ '#e6f9ff','#ccf4ff','#b3eeff','#99e9ff','#80e3ff','#66deff','#4dd8ff','#33d3ff','#1acdff','#00c8ff',
          '#00b4e6','#00a0cc','#008cb3','#007899','#006480','#005066','#00485c','#003c4d','#002833','#00141a']
```

# 1. Spectral radius very close to critical point

```python
# Find the value of n for which the upper branch stability crosses 1, ie when it loses and regains stability
def specrad_cross(t,s,c,n):
    # First find m
    m = opt.root_scalar(lambda x: x/c - 1 + af.fhill(x,n,t),bracket=[c*(n-1)/n,c]).root
    # Now find spectral radius
    return af.specrad(m,s,c,af.fhill,af.dfhill,n,t) - 1
```

```python
# Initial plot to show that this range will include both crossing points
t_range_test = np.linspace(0.97*teq,teq-1e-8,num=1000)
# Note that this works because the end point is teq-1e-8, so the stable point will automatically be included
specm1_test = []
for tt in t_range_test:
    specm1_test.append(specrad_cross(tt,s,c,n))
plt.plot(t_range_test,specm1_test)
plt.axhline(0,c='0.3',zorder=-10)
plt.ylabel('Spectral radius minus 1')
plt.xlabel('Resilience')
```

```python
# Now find exact crossing points
tcross1 = opt.root_scalar(lambda x: specrad_cross(x,s,c,n),bracket=[0.99*teq,teq-1e-8]).root
#[0.606286,0.606286143]) # Old hard coded roots
tcross2 = opt.root_scalar(lambda x: specrad_cross(x,s,c,n),bracket=[0.97*teq,0.99*teq]).root
print(tcross2,tcross1,teq)
```

```python
# Have to make two ranges with different precision and then concatenate
start = np.floor(tcross2*1e3)/1e3 # Point where the stability is returned
mid = np.floor(tcross1*1e5)/1e5 # Middle point where this branch is definitiely unstable
spacing = 1e-8 # Spacing for near bifurcation. Given the 1e5 above, this 1e8 should mean O(10^3) points
# Now set up ranges
t_range21 = np.linspace(start,mid,num=500)
t_range1eq = np.arange(mid,teq+spacing,step=spacing)
# Concatenate
t_range = np.append(t_range21,t_range1eq)

# Now for each theta, solve for m. Solve for bottom and top root and ignore m=0, since that is always a solution
meq = np.zeros((len(t_range),2))

# Define an epsilon away from 0 for bracketing.
me = 1e-8

for i,tt in enumerate(t_range[:-1]):
    # As long as we are not at teq, there should be two roots:
    # Bracket the first room from epsilon to the critical value
    meq[i,0] = opt.root_scalar(lambda x: x/c - 1 + af.fhill(x,n,tt),bracket=[me,ms]).root
    # If we have first root we also have second, above the critical value (except exactly at m*)
    meq[i,1] = opt.root_scalar(lambda x: x/c - 1 + af.fhill(x,n,tt),bracket=[ms,c]).root
    
# At teq, we should only have one root, so set both to that value
meq[-1,0] = ms
meq[-1,1] = ms
```

```python
# Now calculate the spectral radius everywhere
spec = np.zeros((len(t_range),3))

for i,tt in enumerate(t_range):
    spec[i,0] = af.specrad(0,s,c,af.fhill,af.dfhill,n,tt)
    spec[i,1] = af.specrad(meq[i,0],s,c,af.fhill,af.dfhill,n,tt)
    spec[i,2] = af.specrad(meq[i,1],s,c,af.fhill,af.dfhill,n,tt)
        
# Now turn these into indices for stability
stableargs = spec<1
```

```python
fig,ax = plt.subplots(1,1,figsize=(3,3))
ax.plot(t_range,spec[:,2],c=c3,label='Upper branch')
ax.plot(t_range,spec[:,1],c=c3,ls='--',label='Lower branch')
ax.plot(t_range,spec[:,0],c=c3,ls='-.',label='Trivial fixed point')
ax.axhline(1.0,c='k',zorder=-10)
ax.set_box_aspect(1)
ax.set_xlim(start,teq+5e-4)
ax.set_ylim(0.96,1.04)
ax.set_xlabel(r'$\theta$, Resilience')
ax.set_ylabel('Spectral radius')
ax.legend()
fig.savefig('figures/spectralradius.pdf',bbox_inches='tight')
```

```python
fig,ax = plt.subplots(1,1,figsize=(3,3))
ax.plot(t_range,spec[:,2],c=c3,label='Upper branch')
ax.set_box_aspect(1)
ax.axhline(1.0,c='k',zorder=-10)
ax.set_xticks([teq-5e-7,teq], labels=[teq-5e-7,teq])
ax.xaxis.set_major_formatter(plt.FormatStrFormatter('%.7f'))
ax.set_xlim(teq-1e-6,teq+1e-7)
ax.set_ylim(0.985,1.015)
ax.set_xlabel(r'$\theta$, Resilience')
ax.set_ylabel('Spectral radius')
ax.legend()
fig.savefig('figures/spectralradius_onlyupper.pdf',bbox_inches='tight')
```

# 2. Simulations near equilibrium

```python
# Get values of all variables at critical point.
# The easiest way to do this is to get b and then get j and m
# Get critical number of beetles so we can start just above or below critical.
bs = c*(n-1)*(1-s)*s**N/((n-1)*(1+s**N-2*s**(N+1))+n*(1-s)*s**N)
# And for the juvenile classes
js = (s**np.arange(N))*bs/s**N
ss = bs/ms
```

```python
# Which t to simulate for?
t_test = [np.round((teq+tcross1)/2,7),t_range[np.argmax(spec[:,2])],
          np.round((tcross1+tcross2)/2,5),np.round(tcross2,3)]
# Print out the spectral radii there:
for tt in t_test:
    print(tt,specrad_cross(tt,s,c,n)+1)
    
# Plot where these simulations occur
fig,axs = plt.subplots(1,2,figsize=(6,3),sharey=True)
for ax in axs:
    ax.set_box_aspect(1)
    ax.plot(t_range,spec[:,2],c=c3,label='Upper branch')
    #ax.plot(t_range,spec[:,1],c=c3,ls='--',label='Lower branch')
    ax.axhline(1.0,c='k',zorder=-10)
    for tt in t_test:
        ax.axvline(tt,c='0.3')
    ax.set_ylim(0.985,1.015)
    ax.set_xlabel(r'$\theta$, Resilience')

axs[0].set_xlim(0.598,teq+0.0005)
axs[1].set_xlim(0.606279,teq+5e-7)
axs[1].set_xticks([0.606281,0.606284,0.606287], labels=[0.606281,0.606284,0.606287])
axs[1].xaxis.set_major_formatter(plt.FormatStrFormatter('%.6f'))

axs[0].set_ylabel('Spectral radius')
#axs[1].legend()

fig.savefig('figures/spectralradius_simulations.pdf',bbox_inches = 'tight')
```

```python
# Plot dominant eigenvalue itself over the same range
# First get roots and separate real and imaginary
roots = np.zeros([len(t_range),2])
for i,tt in enumerate(t_range):
    root_temp = np.roots(af.coef(meq[i,1],s,c,af.fhill,af.dfhill,n,tt)[::-1])
    specarg_temp = np.argmax(np.absolute(root_temp))
    roots[i,0] = root_temp[specarg_temp].real
    roots[i,1] = root_temp[specarg_temp].imag

fig,axs = plt.subplots(1,2,figsize=(6,3),sharey=True)
for ax in axs:
    ax.set_box_aspect(1)
    ax.plot(t_range,roots[:,0],c=c3,label='Real')
    ax.plot(t_range,roots[:,1],'--',c=c3,label='Imaginary')
    ax.set_xlabel(r'$\theta$, Resilience')
    # Plot vertical lines at intersection points
    #int1 = t_range[np.argmin(np.abs(spec[:100,2]-1))]
    #int2 = t_range[len(spec[:,2])-100+np.argmin(np.abs(spec[-100:-1,2]-1))]
    #for tt in [int1,int2]:
    #    ax.axvline(tt,c='0.1',zorder=-10)

axs[0].set_xlim(0.598,teq+0.0005)
axs[1].set_xlim(0.606279,teq+5e-7)
axs[1].set_xticks([0.606281,0.606284,0.606287], labels=[0.606281,0.606284,0.606287])
axs[1].xaxis.set_major_formatter(plt.FormatStrFormatter('%.6f'))

axs[0].set_ylabel('Dominant eigenvalue')
axs[0].legend()

fig.savefig('figures/spectralradius_simulations_realimaginary.pdf',bbox_inches = 'tight')
```

```python
# Run the simulations
# How many steps to run?
nsteps = 4000
# Generate return array
sim = np.zeros([len(t_test),nsteps,N+2])
# Start at critical values
sim[:,0,0:N] = js
sim[:,0,N] = ss
sim[:,0,N+1] = bs
for j,tt in enumerate(t_test):
    for i in np.arange(nsteps-1):
        # No immigrating beetles, hill function
        sim[j,i+1] = af.timestep(sim[j,i],s,N,n,tt)
```

```python
# Plot simulations near bifurcation
fig,axs = plt.subplots(1,4,figsize=(9.5,2.5),constrained_layout=True)
for j,ax in enumerate(axs):
    ax.set_box_aspect(1)
    ax.plot(sim[j,:,N+1]/sim[j,:,N],c=c3)
    ax.axhline(ms,c='k',zorder=-10)
    ax.set_xlim(0,nsteps)
    ax.set_xlabel('Time (Years)')
    ax.annotate(r'$\theta=${:.7f}'.format(t_test[j]),(0.32,0.1),xycoords='axes fraction')
    
axs[0].set_ylim(0.7995,0.8015)
axs[1].set_ylim(-0.01,0.95)
axs[2].set_ylim(-0.01,0.95)
axs[3].set_ylim(0.795,0.89)
axs[0].set_ylabel('$m$, Mean number of beetles')
fig.savefig('figures/simulation_nearbifurcation.pdf',bbox_inches = "tight")
```

```python
# Plot simulations near bifurcation in a 2x2 array rather than in a line
fig,axs = plt.subplots(2,2,figsize=(5.5,5.5),constrained_layout=True)
for i,ax_x in enumerate(axs):
    ax_x[0].set_ylabel('$m$, Mean number of beetles')
    for j,ax in enumerate(ax_x):
        ax.set_box_aspect(1)
        ax.plot(sim[2*i+j,:,N+1]/sim[2*i+j,:,N],c=c3)
        ax.axhline(ms,c='k',zorder=-10)
        ax.set_xlim(0,nsteps)
        ax.set_xlabel('Time (Years)')
        ax.annotate(r'$\theta=${:.7f}'.format(t_test[2*i+j]),(0.32,0.1),xycoords='axes fraction')
    
axs[0,0].set_ylim(0.7995,0.8015)
axs[0,1].set_ylim(-0.01,0.95)
axs[1,0].set_ylim(-0.01,0.95)
axs[1,1].set_ylim(0.795,0.89)

fig.savefig('figures/simulation_nearbifurcation22.pdf',bbox_inches = "tight")
```

```python
# Run a simulation of extinction time for the appendix
eps = np.linspace(-0.001,0.001,101)
t_extinct = t_test[-2]
n_extinct = 8000 # Chosen from tests to be long enough to allow oscillating simulations to go extinct
# Save the time to zero beetles
sim_extinction = np.zeros([len(eps)])
# Start at critical values but perturb according to eps
for j,ee in enumerate(eps):
    sim_temp = np.zeros([n_extinct,N+2])
    sim_temp[0,0:N] = js
    sim_temp[0,N] = ss-ee
    sim_temp[0,N+1] = bs+ee
    for i in np.arange(n_extinct-1):
        # No immigrating beetles, hill function
        sim_temp[i+1] = af.timestep(sim_temp[i],s,N,n,t_extinct)
    if sim_temp[-1,N+1]==0:
        zeroarg_temp = np.argwhere(sim_temp[:,N+1]==0)[0]
    else:
        zeroarg_temp = 0
    sim_extinction[j] = zeroarg_temp
```

```python
fig,ax = plt.subplots(1,1,figsize=(3,3))
ax.set_box_aspect(1)
ax.plot((bs+eps)/(ss-eps),sim_extinction,'o',c=c3)
ax.axvline(ms,c='0.3')
ax.set_ylabel('Time to extinction (Years)')
ax.set_xlabel(r'$m_0$, Initial mean number of beetles')
ax.set_ylim(-50,n_extinct)
#ax.annotate(r'$\theta=${:.5f}'.format(t_extinct),(0.6,0.1),xycoords='axes fraction')
fig.savefig('figures/simulation_nearbifurcation_extinctiontime.pdf',bbox_inches = "tight")
```
